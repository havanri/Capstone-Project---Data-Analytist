#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['Age_Num'] = df['Age'].map(age_map)

# Chuẩn hóa YearsCodePro
df['YearsCodePro_Clean'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
})
df['YearsCodePro_Clean'] = pd.to_numeric(df['YearsCodePro_Clean'], errors='coerce')

df_clean = df[['Age_Num', 'YearsCodePro_Clean']].dropna()

x = df_clean['Age_Num']
y = df_clean['YearsCodePro_Clean']
coef = np.polyfit(x, y, 1)
trend = np.poly1d(coef)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.5, label='Data', color='lightblue')
plt.plot(np.sort(x), trend(np.sort(x)), color='red', label='Trend Line')

plt.title('Age vs Years of Coding Experience with Trend')
plt.xlabel('Age')
plt.ylabel('YearsCodePro')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
# Drop missing values trước
df_plot = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Age']].copy()

# Chuyển đổi kiểu dữ liệu về numeric
df_plot['ConvertedCompYearly'] = pd.to_numeric(df_plot['ConvertedCompYearly'], errors='coerce')
df_plot['JobSatPoints_6'] = pd.to_numeric(df_plot['JobSatPoints_6'], errors='coerce')

# Ánh xạ tuổi về dạng số
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': None
}
df_plot['Age_Num'] = df['Age'].map(age_map)

# Loại bỏ hàng có giá trị thiếu
df_plot = df_plot.dropna()

# Áp dụng IQR để loại bỏ outliers trong ConvertedCompYearly
Q1 = df_plot['ConvertedCompYearly'].quantile(0.25)
Q3 = df_plot['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Lọc dữ liệu trong khoảng IQR
df_plot = df_plot[(df_plot['ConvertedCompYearly'] >= lower_bound) & (df_plot['ConvertedCompYearly'] <= upper_bound)]

# Chuẩn hóa kích thước bong bóng theo tuổi
sizes = (df_plot['Age_Num'] - df_plot['Age_Num'].min()) / (df_plot['Age_Num'].max() - df_plot['Age_Num'].min()) * 1000

# Vẽ biểu đồ bubble plot
plt.figure(figsize=(15, 8))
scatter = plt.scatter(
    df_plot['JobSatPoints_6'],
    df_plot['ConvertedCompYearly'],
    s=sizes,
    alpha=0.5,
    c=df_plot['Age_Num'],
    cmap='viridis',
    edgecolor='k',
    linewidth=0.5
)

plt.colorbar(scatter, label='Age')
plt.xlabel('Job Satisfaction (1–100)')
plt.ylabel('Yearly Compensation (USD)')
plt.title('Bubble Plot of Compensation vs Job Satisfaction (After IQR Filtering)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
df_lang = df[['LanguageHaveWorkedWith', 'JobSatPoints_6']].copy()
df_lang['JobSatPoints_6'] = pd.to_numeric(df_lang['JobSatPoints_6'], errors='coerce')
df_lang = df_lang.dropna()

# Tách các ngôn ngữ thành từng dòng
df_lang = df_lang.assign(Language=df_lang['LanguageHaveWorkedWith'].str.split(';')).explode('Language')
df_lang = df_lang.dropna(subset=['Language'])

# Lấy top 15 ngôn ngữ phổ biến nhất
top_langs = df_lang['Language'].value_counts().nlargest(15).index
df_lang = df_lang[df_lang['Language'].isin(top_langs)]

# Gán thứ tự x-axis
df_lang['Language'] = pd.Categorical(df_lang['Language'], categories=top_langs, ordered=True)

# Tạo jitter (nhiễu nhẹ) trục X để tránh trùng điểm
df_lang['x_jitter'] = df_lang['Language'].cat.codes + np.random.uniform(-0.2, 0.2, size=len(df_lang))

# Vẽ biểu đồ scatter
plt.figure(figsize=(14, 7))
sns.scatterplot(
    x='x_jitter',
    y='JobSatPoints_6',
    hue='Language',
    data=df_lang,
    alpha=0.4,
    palette='tab10',
    legend=False
)

# Tùy chỉnh trục X
plt.xticks(ticks=range(len(top_langs)), labels=top_langs, rotation=45)
plt.ylabel('Job Satisfaction')
plt.xlabel('Programming Language')
plt.title('Job Satisfaction by Programming Language (using sns.scatterplot)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
df_emp = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Employment']].copy()
df_emp['ConvertedCompYearly'] = pd.to_numeric(df_emp['ConvertedCompYearly'], errors='coerce')
df_emp['JobSatPoints_6'] = pd.to_numeric(df_emp['JobSatPoints_6'], errors='coerce')
df_emp = df_emp.dropna()

# Lọc các Employment phổ biến (tuỳ chọn)
top_employment_types = df_emp['Employment'].value_counts().nlargest(6).index
df_emp = df_emp[df_emp['Employment'].isin(top_employment_types)]

# Áp dụng IQR filter để loại bỏ outlier trong thu nhập
Q1 = df_emp['ConvertedCompYearly'].quantile(0.25)
Q3 = df_emp['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_emp = df_emp[(df_emp['ConvertedCompYearly'] >= lower_bound) & (df_emp['ConvertedCompYearly'] <= upper_bound)]

# Vẽ biểu đồ
plt.figure(figsize=(20, 7))
sns.scatterplot(
    data=df_emp,
    x='JobSatPoints_6',
    y='ConvertedCompYearly',
    hue='Employment',
    style='Employment',       # Dùng marker khác nhau cho từng loại Employment
    alpha=0.6,
    palette='Set2'
)

# Tuỳ chỉnh biểu đồ
plt.title('Job Satisfaction vs Yearly Compensation by Employment Type')
plt.xlabel('Job Satisfaction')
plt.ylabel('Yearly Compensation (USD)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
df_exp = df[['YearsCodePro', 'Age', 'Country']].copy()
df_exp['YearsCodePro'] = pd.to_numeric(df_exp['YearsCodePro'], errors='coerce')
df_exp = df_exp.dropna()

# Lọc top 7 quốc gia phổ biến nhất để dễ quan sát
top_countries = df_exp['Country'].value_counts().nlargest(7).index
df_exp = df_exp[df_exp['Country'].isin(top_countries)]

# Lọc các nhóm tuổi có ý nghĩa (loại 'Prefer not to say' nếu cần)
valid_ages = [
    'Under 18 years old', '18-24 years old', '25-34 years old',
    '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older'
]
df_exp = df_exp[df_exp['Age'].isin(valid_ages)]

# Vẽ biểu đồ
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=df_exp,
    x='Age',
    y='YearsCodePro',
    hue='Country',
    style='Age',
    palette='tab10',
    alpha=0.7
)

# Tuỳ chỉnh biểu đồ
plt.title('Professional Coding Experience by Age Group and Country')
plt.ylabel('Years of Professional Coding (YearsCodePro)')
plt.xlabel('Age Group')
plt.xticks(rotation=30)
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Country')
plt.tight_layout()
plt.show()
```
